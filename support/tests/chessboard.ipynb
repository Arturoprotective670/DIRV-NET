```python
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets
```

```python
def draw_image(img):
    plt.axis("off")
    plt.imshow(img, cmap="gray")
    plt.show()
```

```python
def draw_slice(img, z):
    draw_image(img[:, :, int(z), :])
```

```python
side_size = 4
image_size = 64
img = gradient_chessboard(
    output_size=[image_size] * 3,
    cells_size=[side_size] * 3,
    corners_even_colors=[[23], [123], [154], [15], [177], [200], [34], [78]],
    corners_odd_colors=[[123], [34], [154], [15], [177], [231], [120], [78]],
)
draw_slice(img, 0)
```

```python
slider = widgets.FloatSlider(
    value=2, min=2, max=image_size - 1, step=side_size, description="Slice:"
)  # we jump between middles of the chessboard cubes

interact(lambda z: draw_slice(img, z), z=slider)
```
